# Fontainebleau Validation Protocol

This notebook records the independent Fontainebleau sandstone validation workflow used to test extrapolation to porosity values not directly included as training targets.

```python
import pandas as pd

fontainebleau = pd.DataFrame({
    "sample": ["FB-1", "FB-2", "FB-3", "FB-4"],
    "porosity": [0.1743, 0.1263, 0.0853, 0.2045],
    "shape": ["480x480x480"] * 4,
    "voxel_size_um": [5.68] * 4,
})
fontainebleau
```

## Default Protocol

- Train on the 0.2045 porosity sample by default.
- Generate at 0.2045, 0.1743, 0.1263, and 0.0853.
- Treat lower porosity targets as extrapolation tests.
- Evaluate using porosity, S2, lineal path, EDT, topology, pore-network metrics, and permeability when dependencies are available.

```python
train_command = """
python scripts/train_fontainebleau.py --stage all \\n  --raw_path data/fontainebleau/fontainebleau_phi0p2045.raw \\n  --raw_shape 480 480 480 \\n  --save_dir outputs/fontainebleau_phi0p2045 \\n  --poro_center 0.2045 \\n  --target_porosity 0.2045 \\n  --device cuda
"""
print(train_command)
```

```python
generate_command = """
python scripts/generate_batch.py \\n  --ckpt_dir outputs/fontainebleau_phi0p2045 \\n  --out_root data/generated_fontainebleau_sets \\n  --targets 0.2045 0.1743 0.1263 0.0853 \\n  --n_per_target 50 \\n  --poro_center 0.2045 \\n  --device cuda
"""
print(generate_command)
```
